1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

Accuracy: The accuracy metric tells us how many of the results are did the model predict correctly (i.e. when the model predicted a negative, the result was truly a negative). This metric is useful in weather forecasting by determining how many times you correctly predict rain (i.e. it rains when you predict it will and does not rain when you predict it won't). Accuracy is useful because there is likely a small class imbalance in weather networking (i.e. 30% rain vs 70% rain rather than 1% rain vs 99% rain). A low class imbalance is crucial for accuracy to be useful because accuracy favors the majority class. Accuracy is also useful when the consequences of incorrect predictions for both classes (majority and minority) are similar.

Sensitivity: The sensitivity measure tells us how many positives we correctly identified (i.e. we predicted a positive and it was truly a positive and we didn't miss many positives). This is useful in HIV testing. In this case, we want to maximize sensitivity because missing a positive case will allow the HIV disease to spread. Sensitivity is more useful than specificity in this case because if a patient is incorrectly identified to have HIV when they do not (false positive), that patient does not pose a risk to society so while undesired, a false positive can be managed by taking another test. Thus sensitivity is specifically useful when we must ensure false negatives are eliminated.

Specificity: The specificity metric tells us the proportion of actual negatives that we identified. That is we predicted a negative and it as a negative and we didn't miss many negatives. This metric is useful in a spam email detection. In this case, we do not want to falsely identify important emails as spam (a false positive) and delete them. Doing so could cause us to lose important information. Specificity is more important than sensitivity here because identifying emails as not spam when they are can be fixed by deleting them later. 

Prediction: Prediction focuses on the proportion of positive predictions that are actually positives. That is, we predicted a positive *correctly*. This is important in targeting online marketing campaign. A company would want to predict which customers are likely to click on a advertisements and the company wants to pick the advertisements tailored to the customer. This is focused on precision because if we predict the customer would buy the item when they do not, we will be wasting marketing budget on a false positive. Ultimately, precision is useful when false positives have significant costs.

### ANSWER:
The 2 lines given in the question prepare the data for the decision tree classifier model. The target variable y is created by converting the categorical variable Hard_or_Paper into a binary variable by using .get_dummies. This encodes each value (hard or paper) into 0 or 1 where 1 represents 'Hardcover'. 

The independent variable X selected as the column ['List Price']. This means that model predicts whether the book is hard cover or paper back depending on the price of the book.

Furthermore, this model should be trained on a training data which is specifically designated for the model. The test data is used later to determine whether the model works as well on unseen data and must therefore remain separted from the training data. This is also why ab_reduced_noNaN is not used as using this to train would mean there is no data to test the model.

````python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Create an 80/20 split
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=42)

# Report the number of observations in each set
train_size = ab_reduced_noNaN_train.shape[0]
test_size = ab_reduced_noNaN_test.shape[0]

print(f"Training set size: {train_size} observations")
print(f"Test set size: {test_size} observations")

# Separate features and target for training and testing data for clf (List Price only)
X_train_clf = ab_reduced_noNaN_train[['List Price']]  # Features (List Price) from training set
y_train_clf = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  # Target (Hardcover binary) from training set

X_test_clf = ab_reduced_noNaN_test[['List Price']]  # Features (List Price) from test set
y_test_clf = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # Target (Hardcover binary) from test set

# Train the Decision Tree Classifier using training data for clf
clf = DecisionTreeClassifier(max_depth=2, random_state=42)  # Instantiate with max depth 2
clf.fit(X_train_clf, y_train_clf)  # Train the classifier on training data

# Visualize the decision tree for clf
plt.figure(figsize=(10, 6))
tree.plot_tree(clf, 
               feature_names=['List Price'],  # Name of the feature(s)
               class_names=['Paperback', 'Hardcover'],  # Name of the target classes
               filled=True,  # Fill nodes with color for clarity
               rounded=True)  # Round the corners for aesthetics
plt.show()

````

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**.
